```python
## 4.1 Heatmap: Hospitals vs Years
pivot = df1.pivot_table(index="HospitalName", columns="year", values="Total", aggfunc="sum").fillna(0)
plt.figure(figsize=(12,8))
sns.heatmap(pivot, cmap="Reds", linewidths=0.5)
plt.title("Heatmap: Total Patients by Hospital and Year")
plt.show()
```

```python
## 4.3 Waiting Bands Trend Over Years
bands_year = df1.groupby("year")[["0-6 Months","6-12 Months","12-18 Months","18 Months +"]].sum()
bands_year.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Waiting Time Bands Over Years")
plt.ylabel("Patients")
plt.show()
```
